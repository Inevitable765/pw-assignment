```python
"""2. Tasks to Be Performed:


1. Create a Git working directory with feature1.txt and feature2.txt in the master branch

2. Create 3 branches develop, feature1 and feature2

3. In develop branch create develop.txt, do not stage or commit it

4. Stash this file and check out to feature1 branch

5. Create new.txt file in feature1 branch, stage and commit this file

6. Checkout to develop, unstash this file and commit

7. Please submit all the Git commands used to do the above step>"""

1.mkdir gittest #creating a directory by mkdir
  cd gittest  # going to the gittest directory
  git init  #initialize a repo
  touch feature1.txt feature2.txt  #creating file with touch
  git add .  # this will add all the files
  git commit -m "commit to master "  #commits the changes


2.git branch develop   #create a new branch
  git branch feature1  #new branch
  git branch feature2  #new branch


3.git checkout develop   #switch the branch
  touch develop.txt


4.git stash            #stash a untracked file
  git checkout feature1   #switch the branch



5.touch new.txt        # Create new.txt file
 git add new.txt      # Stage the file
 git commit -m "commit to branch"  # Commit the file


6.git checkout develop  # Switch back to develop branch
 git stash pop        # Apply the stashed changes (develop.txt)
 git add develop.txt  # Stage develop.txt
 git commit -m "commit the stash file"  # Commit the file


 7.git log --oneline --all --graph  # View commit history
git branch  # List all branches
git status  # Check the current status


```

```python
"""3. Create a Git working directory, with the following branches:


               Develop

               F1

               f2


2. In the master branch, commit main.txt file

3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively

4. Push all these branches to GitHub

5. On local delete f2 branch

6. Delete the same branch on GitHub as well"""



1.mkdir GitProject  # Create a new directory
cd GitProject     # Navigate into the directory
git init          # Initialize a Git repository




2.git branch develop  # Create develop branch
git branch F1       # Create F1 branch
git branch f2



3.touch main.txt     # Create main.txt
git add main.txt   # Stage the file
git commit -m "commit main.txt"  # Commit the file



4.git checkout develop  # Switch to develop branch
touch develop.txt      # Create develop.txt
git add develop.txt    # Stage the file
git commit -m "commit develop"  # Commit




git checkout F1       # Switch to F1 branch
touch f1.txt          # Create f1.txt
git add f1.txt        # Stage the file
git commit -m "commit f1"  # Commit


git checkout f2       # Switch to f2 branch
touch f2.txt          # Create f2.txt
git add f2.txt        # Stage the file
git commit -m "commit f2"  # Commit




5.git remote add origin https://github.com/Inevitable765/pw-assignment.git  # GitHub repository URL
git push -u origin master  # Push master branch
git push -u origin develop  # Push develop branch
git push -u origin F1  # Push F1 branch
git push -u origin f2  # Push f2 branch



6.git branch -d f2  # Delete f2 branch

git push origin --delete f2  # Delete the same branch on GitHub
```

```python
"""4. Tasks to Be Performed:

1. Put master.txt on master branch, stage and commit

2. Create 3 branches: public 1, public 2 and private

3. Put public1.txt on public 1 branch, stage and commit

4. Merge public 1 on master branch

5. Merge public 2 on master branch

6. Edit master.txt on private branch, stage and commit

7. Now update branch public 1 and public 2 with new master code in private

8. Also update new master code on master

9. Finally update all the code on the private branch """


1.touch master.txt       # Create master.txt file
git add master.txt     # Stage the file
git commit -m "commit master"  # Commit the file


2.git branch "public 1"   # Create public 1 branch
git branch "public 2"   # Create public 2 branch
git branch private      # Create private branch


3.git checkout "public 1" # Switch to public 1 branch
touch public1.txt       # Create public1.txt
git add public1.txt     # Stage the file
git commit -m "commit public1"  # Commit the file


4.git checkout master     # Switch  branch
git merge "public 1"    # Merge public


5.git merge "public 2"    # Merge public 2


6.git checkout private  # Switch to private branch
echo "Updated " >> master.txt  # Edit master.txt
git add .    # Stage the changes
git commit -m "message"  # Commit the changes


7.git checkout "public 1"  # Switch to public 1 branch
git merge private        # Merge private branch into public 1 branch

git checkout "public 2"  # Switch to public 2 branch
git merge private


8.git checkout private  # Switch to private branch

9.git merge master      # Merge master branch into private
```
